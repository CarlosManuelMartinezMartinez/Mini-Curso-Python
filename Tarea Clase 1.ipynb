# Tarea 1
Programa en Python que solicite al usuario una calificación numérica entre 0 y 10.
- 1. Pedir la calificación usando input().
- 2. Convertir el valor ingresado a tipo float. 
- 3. Imprimir la clasificación correspondiente. Si la calificación es menor que 0 o mayor que 10, imprimir: "Calificación no válida".

```python
# Introduce una calificación.
calif=float(input("Introduce la calificación: "))

if calif>= 8.6 and calif<=10:
    print("Tu calificación ",calif,"corresponde a MB.")
elif calif>= 7.6 and calif<8.6:
    print("Tu calificación ",calif,"corresponde a B.")
elif calif>= 6 and calif<7.6:
    print("Tu calificación ",calif,"corresponde a S.")
elif calif>= 0 and calif<6:
    print("Tu calificación ",calif,"corresponde a NA.")
else:
    print("Calificación no válida")
```

# Tarea 2

Análisis de una lista de números.  
Programa que:
- Imprima todos los elementos de la lista usando un ciclo for.
- Calcule la suma de todos los elementos. 
- Calcule cuántos números son pares. 
- Calcule cuántos números son impares.
- Construya una nueva lista llamada mayores_20 que contenga solamente los números mayores que 20.
- Imprima los resultados.

```python
# Lista inicial
numeros = [4, 7, 12, 15, 18, 21, 24, 30, 33, 40]
mayores_20=[]
suma=0
contador=0
contador_par=0
contador_impar=0
cond=True

for i in numeros:
    contador=contador+1
    suma=suma+i
    cond=True
    
    #Contar pares e impares
    if i%2==0:
        contador_par=contador_par+1
    else:
        cond=False
        contador_impar=contador_impar+1
    
    
    #Imprimir resultados
    if cond:
        print(contador,i,"Par")
    else:
        print(contador,i,"Impar")

    #Lista mayores_20
    if i>20:
        mayores_20.append(i)

print("\nElementos: ",contador," Pares: ",contador_par, " Impares: ",contador_impar," Suma: ",suma,"\nLista de mayores a 20: ",mayores_20)
```
