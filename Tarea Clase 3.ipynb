## Tarea 1
Simula el lanzamiento de dos dados juntos durante 1000 repeticiones. Realiza:
- Utiliza paqueteria Random
- Guarda en una lista la suma de los dos dados en cada lanzamiento.
- Cuenta cuántas veces aparece cada suma posible, desde 2 hasta 12.

```python
pip install matplotlib
```

```python
import random
import matplotlib.pyplot as plt
# Simular lanzamientos

#Guarda en una lista la suma de los dos dados en cada lanzamiento.
n=1000
lista=[]
for i in range(n):
    dado1=random.randint(1,6) #Entero aleatorio entre 1 y 6
    dado2=random.randint(1,6)
    lista.append(dado1+dado2)
print(lista)
#print(lista)
#Cuenta cuántas veces aparece cada suma posible, desde 2 hasta 12.
contador_2=0;contador_3=0;contador_4=0;contador_5=0;contador_6=0;contador_7=0;contador_8=0;contador_9=0;contador_10=0;contador_11=0;contador_12=0
for i in lista:
    if i==2:
        contador_2=contador_2+1
    elif i==3:
        contador_3=contador_3+1
    elif i==4:
        contador_4=contador_4+1
    elif i==5:
        contador_5=contador_5+1
    elif i==6:
        contador_6=contador_6+1
    elif i==7:
        contador_7=contador_7+1
    elif i==8:
        contador_8=contador_8+1
    elif i==9:
        contador_9=contador_9+1
    elif i==10:
        contador_10=contador_10+1
    elif i==11:
        contador_11=contador_11+1
    else:
        contador_12=contador_12+1

print("Contadores: \nContador 2: ",contador_2,"\nContador 3: ",contador_3,"\nContador 4: ",contador_4,"\nContador 5: ",contador_5,"\nContador 6: ",contador_6)
print("Contador 7: ",contador_7,"\nContador 8: ",contador_8,"\nContador 9: ",contador_9,"\nContador 10: ",contador_10,"\nContador 11: ",contador_11,"\nContador 12: ",contador_12)

#Calcula la frecuencia relativa de obtener una suma igual a 7.
print("El contador para 7: ",contador_7)
print("La frecuencia relativa de obtener una suma igual a 7 es: ",contador_7/n," Corresponde al ",contador_7/n*100,"% de los datos.")

#Grafica un histograma de las sumas obtenidas.
plt.hist(lista,bins=range(2, 13),edgecolor="black")
plt.xticks(range(2, 13)) 
plt.xlabel("Suma de Dados")
plt.ylabel("Frecuencia")
plt.title("Frecuencia de resultados de la suma de ambos dados por repetición")
plt.show()


```

### ¿Por qué la suma 7 aparece con más frecuencia que la suma 2 o la suma 12?

La distribución corresponde a una aproximadamente normal. A números grandes tiende a distribución normal.
