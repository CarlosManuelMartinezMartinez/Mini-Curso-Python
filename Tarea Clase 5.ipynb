```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
ruta2020="https://raw.githubusercontent.com/jamc88/min-curso/refs/heads/main/Datos/nacimientos_2020.csv"
ruta2021="https://raw.githubusercontent.com/jamc88/min-curso/refs/heads/main/Datos/nacimientos_2021.csv"
ruta2022="https://raw.githubusercontent.com/jamc88/min-curso/refs/heads/main/Datos/nacimientos_2022.csv"

archivos=[(ruta2020,2020),(ruta2021,2021),(ruta2022,2022)]  # se accede usando archivos[i][j]
bases=[]
for ruta,año in archivos:
    df=pd.read_csv(ruta)
    df["Año"]=año
    bases.append(df)
#print(bases)  es un una lista de dataframes
df_nacimientos=pd.concat(bases,ignore_index=True)

df_nacimientos
```

```python
nacim=df_nacimientos.set_index("Año")

for i in range(2020,2023):
    nacimientos=nacim.loc[i]["Total"].sum()
    print("Los nacimientos de ",i," son: ",nacimientos)

# de forma alternativa dada como serie
nacimientos=df_nacimientos.groupby("Año")["Total"].sum()
nacimientos
```
