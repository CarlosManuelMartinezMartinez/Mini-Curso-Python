# Tarea 1
Resuelve el siguiente sistema de ecuaciones aplicando el método de Cramer:

$$
\begin{align}
2x_{1} + 4x_{2} + 6x_{3} &= 18 \\
4x_{1} + 5x_{2} + 6x_{3} &= 24 \\
3x_{1} + x_{2} - 2x_{3} &= 4
\end{align}
$$


```python
# Resolver sistema por método de Cramer
import numpy as np
A=np.array([ [2,4,6],[4,5,6],[3,1,-2] ])
b=np.array([ [18],[24],[4] ])
D=np.linalg.det(A) 
#print(A)

if D!=0:
    #Calcular x1
    A1=A.copy()
    A1[:,0]=b[:,0]
    D1=np.linalg.det(A1)
    x1=D1/D
    
    #Calcular x2
    A2=A.copy()
    A2[:,1]=b[:,0]
    D2=np.linalg.det(A2)
    x2=D2/D
    
    #Calcular x3
    A3=A.copy()
    A3[:,2]=b[:,0]
    D3=np.linalg.det(A3)
    x3=D3/D

    print("Los valores son :",x1,x2,x3)
else:
    print("El sistema no tiene solución o tiene infinitas soluciones")
```

Importa NumPy 
Escribe un programa que haga lo siguiente:

- 1. Defina una función llamada distancia(u, v) que regrese la distancia euclidiana entre dos vectores.
- 2. Defina una función llamada normalizar(x) que regrese el vector normalizado.
- 3. Calcule el producto escalar entre u y v usando np.dot().
- 4. Calcule la norma de u y la norma de v.
     Imprime:
     - La distancia entre u y v.
     - El vector u normalizado.
     - El vector v normalizado.
     - El producto escalar.

```python
import numpy as np
import math

#Vectores
u=[1,2,5]
v=[2,4,5]

# 1. Defina una función llamada distancia(u, v) que regrese la distancia euclidiana entre dos vectores.
def distancia(u,v):
    suma=0
    for i in range(len(u)):
        suma=suma+(v[i]-u[i])**2
    return math.sqrt(suma)

print("La distancia entre u y v es: ",distancia(u,v))

# 2. Defina una función llamada normalizar(x) que regrese el vector normalizado.
def normalizar(x):
    suma=0
    for i in range(len(x)):
        suma=suma+(x[i])**2
    copia=x[:] #copia de vector
    for i in range(len(x)):
        copia[i]=x[i]/math.sqrt(suma)
    return copia

# 3.  Calcule el producto escalar entre u y v usando np.dot()
print("El valor del producto escalar es: ",np.dot(u,v))

# 4. Calcular la norma de u y v
print("El vector u normalizado: ",normalizar(u))
print("El vector v normalizado: ",normalizar(v))
```
